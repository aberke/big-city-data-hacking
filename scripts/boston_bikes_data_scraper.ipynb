```python
"""
Downloads the zip / csv files listed on https://s3.amazonaws.com/hubway-data/index.html
"""

import pandas as pd
import requests

ARCHIVE_URL = "https://s3.amazonaws.com/hubway-data/"

# Make the list of filenames and then download them
filenames = []
# From 01-2015 to 04-2018 they look like this: 201804-hubway-tripdata.zip
# From 05-2018 to (present) 07-2019 they look like this: 201805-bluebikes-tripdata.zip, 201907-bluebikes-tripdata.zip
for year in range(2015, 2023):
    for month in range(1, 13):
        filename = "%d%s-hubway-tripdata.zip" % (year, str(month).zfill(2))
        if (year == 2018 and month > 4) or (year > 2018):
            filename = "%d%s-bluebikes-tripdata.zip" % (year, str(month).zfill(2))
        filenames.append(filename)

    
# And then there are these files with other names
filenames += [
    "hubway_Trips_2011.csv",
    "hubway_Trips_2012.csv",
    "hubway_Trips_2013.csv",
    "hubway_Trips_2014_1.csv",
    "hubway_Trips_2014_2.csv",
]

for filename in filenames:
    url = ARCHIVE_URL + filename
    print("Downloading %s" % url)
    resp = requests.get(url)
    if resp.status_code != 200:
        print("%s not found, skipping! (status: %d)" % (filename, resp.status_code))
        continue

    with open("../data/boston-bike/%s" % filename, 'wb') as f:
        f.write(resp.content)
```
