```python
import requests

ARCHIVE_URL = "https://s3.amazonaws.com/capitalbikeshare-data/"


def do_http_request(url, filename):
    response = requests.get(url)

    if response.status_code != 200:
        print(f'{filename} not found, skipping (status {response.status_code}')
        return


    with open(f'../data/dc-bike/{filename}', 'wb') as f:
        f.write(response.content)
        



single_years = range(2010, 2018)

monthly_years = range(2018, 2023)
months = range(1, 13)


for year in single_years:
    filename = f'{year}-capitalbikeshare-tripdata.zip'
    url = ARCHIVE_URL + filename
    print(f'Downloading {url}')
    do_http_request(url, filename)

for year in monthly_years:
    for month in months:
        filename = f'{year}{str(month).zfill(2)}-capitalbikeshare-tripdata.zip'
        url = ARCHIVE_URL + filename
        print(f'Downloading {url}')
        do_http_request(url, filename)

    

```
